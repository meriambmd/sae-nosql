```python
# Importation des modules utilisés
import sqlite3
import pandas
```

```python
# Création de la connexion
conn = sqlite3.connect("ClassicModel.sqlite")
```

```python
#1 Lister les clients n’ayant jamais effecuté une commande
pandas.read_sql_query("SELECT * FROM Customers c LEFT JOIN Orders o ON c.customernumber = o.customernumber WHERE ordernumber IS NULL;", conn)
```

```python
#### 2 Pour chaque employé, le nombre de clients, le nombre de commandes et le montant total de celles-ci
pandas.read_sql_query("""SELECT e.employeeNumber, e.lastName, e.firstname, count(DISTINCT c.customernumber) as 'Nb de clients', count(DISTINCT o.ordernumber) AS 'Nb commande', sum(p.amount) AS 'Montant total'
FROM Employees e 
LEFT JOIN Customers c ON c.salesRepEmployeeNumber = e.employeeNumber 
LEFT JOIN Orders o ON o.customernumber = c.customernumber 
LEFT JOIN Payments p ON p.customernumber= c.customernumber 
GROUP BY 1 
ORDER BY 4 DESC;""", conn)
```

```python
#### 3 Idem pour chaque bureau (nombre de clients, nombre de commandes et montant total), avec en plus le nombre de clients d’un pays différent, s’il y en a ;
pandas.read_sql_query("""SELECT of.officecode, count(DISTINCT c.customernumber) as 'Nb de clients', count(DISTINCT o.ordernumber) AS 'Nb commande', sum(p.amount) AS 'Montant Total', 
    COUNT(DISTINCT CASE 
                WHEN of.country != c.country 
                THEN c.customernumber 
                ELSE NULL 
            END) AS 'Clients Pays différents'
FROM Employees e 
LEFT JOIN Customers c ON c.salesRepEmployeeNumber = e.employeeNumber 
LEFT JOIN Orders o ON o.customernumber = c.customernumber 
LEFT JOIN Payments p ON p.customernumber= c.customernumber 
LEFT JOIN Offices of ON e.officecode = of.officecode
GROUP BY 1 
ORDER BY 2 DESC;""", conn)
```

```python
#4 Pour chaque produit, donner le nombre de commandes, la quantité totale commandée, et le nombre de clients différents ; 
pandas.read_sql_query("""SELECT p.productcode, p.productName, count(o.ordernumber) as 'Nb commmandes', sum(od.quantityordered) as 'Qte total', count(DISTINCT c.customernumber) as 'Nb client'
FROM Products p 
LEFT JOIN orderdetails od ON od.productcode = p.productcode 
LEFT JOIN Orders o ON o.ordernumber = od.ordernumber 
LEFT JOIN Customers c ON c.customernumber = o.customernumber
GROUP BY 2
ORDER BY 3 DESC;""", conn)
```

```python
#5.	Donner le nombre de commande pour chaque pays du client, ainsi que le montant total des commandes et le montant total payé : on veut conserver les clients n’ayant jamais commandé dans le résultat final 
pandas.read_sql_query("""SELECT c.country, count(o.ordernumber) as 'Nb de commande', sum(quantityordered*priceeach) as 'Montant total', sum(p.amount) AS 'Montant Total Payé'
FROM Customers c
LEFT JOIN Orders o ON o.customernumber = c.customernumber
LEFT JOIN Payments p ON c.customernumber = p.customernumber
LEFT JOIN orderdetails od ON od.ordernumber = o.ordernumber 
GROUP BY 1
ORDER BY 2 DESC;""", conn)
```

```python
#6.	On veut la table de contigence du nombre de commande entre la ligne de produits et le pays du client 
pandas.read_sql_query("""SELECT P.productLine, C.country , COUNT(O.orderNumber) AS 'Nb commande'
FROM Products P
JOIN OrderDetails OD ON P.productCode = OD.productCode
JOIN Orders O ON OD.orderNumber = O.orderNumber
JOIN Customers C ON O.customerNumber = C.customerNumber
GROUP BY 1;""", conn)
```

```python
#7.	On veut la même table croisant la ligne de produits et le pays du client, mais avec le montant total payé dans chaque cellule  
pandas.read_sql_query("""SELECT P.productLine, C.country, SUM(P.amount) AS 'Montant Total Payé'
FROM Products P
JOIN OrderDetails OD ON P.productCode = OD.productCode
JOIN Orders O ON OD.orderNumber = O.orderNumber
JOIN Customers C ON O.customerNumber = C.customerNumber
JOIN Payments P ON C.customerNumber = P.customerNumber
GROUP BY 1;""", conn)
```

```python
#8.	Donner les 10 produits pour lesquels la marge moyenne est la plus importante (cf buyPrice et priceEach) ;
pandas.read_sql_query("""SELECT P.productCode, P.productName, AVG(OD.priceEach - P.buyPrice) AS 'Marge Moy'
FROM Products P
JOIN OrderDetails OD ON P.productCode = OD.productCode
GROUP BY 1
ORDER BY 3 DESC
LIMIT 10;""", conn)
```

```python
#9.	Lister les produits (avec le nom et le code du client) qui ont été vendus à perte :
#o	Si un produit a été dans cette situation plusieurs fois, il doit apparaître plusieurs fois,
#o	Une vente à perte arrive quand le prix de vente est inférieur au prix d’achat ;
pandas.read_sql_query("""SELECT P.productCode, P.productName, C.customernumber, C.customerName, OD.priceEach, P.buyPrice
FROM Products P
JOIN OrderDetails OD ON P.productCode = OD.productCode
JOIN Orders O ON OD.orderNumber = O.orderNumber
JOIN Customers C ON O.customerNumber = C.customerNumber
WHERE OD.priceEach < P.buyPrice;""", conn)
```

```python
#10. Lister les clients pour lesquels le montant total payé est inférieur aux montants totals des achats ;
pandas.read_sql_query("""SELECT C.customerNumber, C.customerName, 
       SUM(P.amount) AS totalPaid, 
       SUM(OD.quantityOrdered * OD.priceEach) AS totalOrdered
FROM Customers C
JOIN Orders O ON C.customerNumber = O.customerNumber
JOIN Payments P ON C.customerNumber = P.customerNumber
JOIN OrderDetails OD ON O.orderNumber = OD.orderNumber
GROUP BY 1
;""", conn)
```
